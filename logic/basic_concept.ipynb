# 基本概念

## 命题
- 陈述事实的句子P。
- 句子能判断真假
- 句子要么真，要假。不能既真又假。
- 有限个命题通过有限种运算符进行有限次组合后还是命题

### 父概念
- 句子
- 真
- 假
- 有限
- 命题运算，包括运算方式，组合规则


### 约束
- 陈述事实
- 能判断
- 有确定唯一的真假状态

## 真值表
- 表格中的值只有真和假两种的表格
- 对表格的列数和行数没有限制，举例如下：
|P1|P2|
|:----|:----|
|真|假|
|假|真|

### 父概念
- 表格
- 表格的取值
- 真
- 假

### 约束
- 表格的取值只能是真，假这两个值。

## 命题运算定义

### 非
规定运算符号：$\neg$
- 非真等于假 $\neg True = False$
- 非假等于真 $\neg False = True$
- 命题P可能是真，可能是假，取不同值时非P对应如下：
|$P$|$\neg P$|
|:----|:----|
|True|False|
|False|True|

#### 父概念
- 真 True
- 假 False
- 等于 =
- 命题
- 真值表

#### 约束
- 等于 =

### 单条件： 蕴含
如果A那么B，若A则B，A蕴含B
- 规定符号：$A \rightarrow B$
- A, B是命题
- 真值表

### 双条件： 当前仅当
如果A那么B 析取 如果B那么A。 A当前仅当B
- 规定符号：$A \leftrightarrow B$
- A, B是命题
- 真值表

## 谓词
- 命题语句(主谓宾)中的谓语和宾语叫作谓词P。
- 主语可变记作x
- 这个命题可表示为P(x)

## 量词
### 全称量词
- 对某限定集合S中的每个元素x，命题P(x)为真: $\forall (x \in S) P(X)$

### 存在量词
- 对某限定集合S中的存在一个或者多个元素x，命题P(x)为真: $\exists (x \in S)  P(X)$

### 唯一量词
- 对某限定集合S中的存在一个且仅有一个元素x，命题P(x)为真: $\exists ! (x \in S)  P(X)$

#### 父概念
- 集合
- 集合元素
- 属于
- 一个， 多个， 每个

## 摩根律
- $\neg\exists xP(x) \leftrightarrow \forall x \neg P(x)$
- $\neg \forall xP(x) \leftrightarrow \exists x \neg P(x)$

## 推理规则

### 论证

生成一个命题序列，P1,P2,P3...Pn, 其中Pn为结论， P1至Pn为前提。 这个过程称为论证

- 认证的有效性
Pn在P1至Pn-1的所有命题为真时必为真。 $(P1 \wedge P2 \wedge P3 ... \wedge Pn-1) \rightarrow Pn$
